## Iris dataset

##### This is perhaps the best known database to be found in the pattern recognition literature. The Iris flower data set or Fisher's Iris data set is a multivariate data set introduced by the British statistician and biologist Ronald Fisher in his 1936 paper The use of multiple measurements in taxonomic problems as an example of linear discriminant analysis.

![Iris flowers](http://python.astrotech.io/_images/iris-flowers.png)
![Scatterplot of UCI dataset](https://upload.wikimedia.org/wikipedia/commons/thumb/5/56/Iris_dataset_scatterplot.svg/220px-Iris_dataset_scatterplot.svg.png)

### Iris dataset import and loading
##### The dataset contains a set of 150 records under five attributes - petal length, petal width, sepal length, sepal width and species.
##### There are no null values.
##### There are 50 observations of each species (setosa, versicolor, virginica).
##### Measurements: sepal length, sepal width, petal length, petal width
##### There is a strong relationship between the measurements and the species. Machine learning models can actually predict the species based on the measurements.

## Plotting with Iris Dataset

### Multivariate Plots
##### Interactions between the variables.
##### Diagonal grouping of some pairs of attributes suggests a high correlation and a predictable relationship.
##### After graphing the features in a pair plot, it is clear that the relationship between pairs of features of a iris-setosa which is blue is distinctly different from those of the other two species.
##### There is some overlap in the pairwise relationships of the other two species, iris-versicolor (brown) and iris-virginica (green).


```python
import seaborn as sns
```

```python
g = sns.pairplot(df, hue='species', markers='+')
plt.show()
```

### Univariate Plots

##### Univariate plots, that is, plots of each individual variable.

##### Given that the input variables are numeric, we can create box and whisker plots of each.

```python
df.plot(kind='box', subplots=True, layout=(2,2), sharex=False, sharey=False) # box and whisker plots
plt.show()
```

```python
df.hist() # datset on histogram
plt.show()
```

### Seaborn Visualization
##### Violinplots allow to visualize the distribution of a numeric variable
##### Violins are particularly adapted when the amount of data is huge and showing individual observations gets impossible. 
##### Gives you all the length and widths of the different species

```python
g = sns.violinplot(y='species', x='sepal_length', data=df, inner='quartile')
plt.show()
```

```python
g = sns.violinplot(y='species', x='sepal_width', data=df, inner='quartile')
plt.show()
```

```python
g = sns.violinplot(y='species', x='petal_length', data=df, inner='quartile')
plt.show()
```

```python
g = sns.violinplot(y='species', x='petal_width', data=df, inner='quartile')
plt.show()
```

### IRIS Correlation Matrix
#####  Correlation matrix is used to see parametrs which best correlate each other
##### According to the correlation matrix results Petal LengthCm and PetalWidthCm have positive correlation which is proved by the scatter plot above

```python
cm = df.corr()
cm
```

```python
plt.figure(figsize=(10,8)) 
sns.heatmap(cm, 
            xticklabels=cm.columns.values,
            yticklabels=cm.columns.values,
           cmap='viridis', annot=True)
plt.show()
```
